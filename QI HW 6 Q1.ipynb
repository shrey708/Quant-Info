```python
from sympy import symbols, exp, integrate, oo

# Define symbols
t, tau = symbols('t tau')

# Define the probability density function
P = (1/tau) * exp(-t/tau)

# Compute the expectation value of t
E_t = integrate(t * P, (t, 0, oo))
E_t
```

```python

n_iterations = 1000  

mean_values = np.zeros(n_iterations)

# Repeat the process of generating samples and calculating their mean 1000 times
for i in range(n_iterations):
    
    samples = np.random.exponential(scale=tau, size=n_samples)

    mean_values[i] = np.mean(samples)

mean_values[:100]  # Show the first 10 mean values to check the output
```

```python
import matplotlib.pyplot as plt

# Create a histogram of the mean values
plt.hist(mean_values, bins=30, edgecolor='black')
plt.title('Histogram of Means from Exponential Distribution Samples')
plt.xlabel('Mean Value')
plt.ylabel('Frequency')

plt.show()
```

The normalization of the histogram allows it to simulate the likelihood function for the mean of an exponential distribution, showing how the probability density of these means is distributed around the true mean $\tau$ of the population.

This visualization effectively illustrates the central limit theorem, indicating that the distribution of sample means approximates a normal distribution centered around the population mean, even when the underlying distribution (in this case, exponential) is not normal.

```python

n_iterations = 1000  
n_samples = 1000  
tau = 1  

mean_values = np.zeros(n_iterations)

# Repeat the process of generating samples and calculating their mean 1000 times
for i in range(n_iterations):
    
    samples = np.random.exponential(scale=tau, size=n_samples)
    
    mean_values[i] = np.mean(samples)

# Create a histogram of the mean values with normalization to simulate the likelihood function
plt.hist(mean_values, bins=30, density=True, edgecolor='black', label='Normalized Histogram')


plt.title('Normalized Histogram of Means (Likelihood Simulation)')
plt.xlabel('Mean Value')
plt.ylabel('Probability Density')

plt.legend()
plt.show()
```
